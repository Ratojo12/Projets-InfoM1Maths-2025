```python
import matplotlib.pyplot as plt
import math
#Definition de la foncton f(x)=a*e^{bx}
def f(x,a,b):
    return a*math.exp(b*x)
#Posons d'abord a=1 et b=3 et tracons la courbe
a=1
b=3
x_valeur=[]
y_valeur=[] 
# Remplir les listes x_valeur et y_valeur en utilisant la boucle for
for x in range(-50,51):  #Correspond à x allant de -5 à 5 
    x_val=x/10      # avec un pas 0.1
    x_valeur.append(x_val)
    y_valeur.append(f(x_val,a,b))

plt.plot(x_valeur,y_valeur)
plt.title("Representation de y = 1*e^{3x}")
plt.xlabel("x")
plt.ylabel("1*e^{3x}")
plt.show()
```

```python
# Approximation en utilisant la série de Taylor
def approximation(x,a,b,n=1000):
    approximation= a *1
    for n in range(1,n):
        approximation += a*((b**n *x**n)/math.factorial(n))
    return approximation

a=1
b=3
x_valeur=[]
y_valeurapprox=[]
#Visualisation de l' approximation
for x in range(-50,50):
    x_val=x/10
    x_valeur.append(x_val)
    y_valeurapprox.append(approximation(x,a,b,n=1000))

plt.plot(x_valeur,y_valeurapprox)
plt.title("Representation de l'approximation de 1*e^{3x}")
plt.xlabel("x")
plt.ylabel("1*e^{3x}")
plt.show()

```

```python
#Analyse visuel et graphique pour différentes valeurs de a et b
x_valeur=[i*0.1 for i in range(-20,21)]#Valeur de x entre -2 et 2 avec n pas de 0.1

def visualiser_croissance(a_valeur,b_valeur):
    x_valeur=[]
    y_valeur=[]
    for x in range(-50,50):
        x_val=x/10
        x_valeur.append(x_val)
        y_valeur.append(f(x_val,a_valeur,b_valeur))
        

    plt.plot(x_valeur,y_valeur)
    plt.title("Representation de y = a*e^{bx}")
    plt.xlabel("x")
    plt.ylabel("a*e^{bx}")
    plt.show()
visualiser_croissance(3,3)
print("La fonction croît très rapidement vers plus l'infinie,la courbe monte de manière exponentielle(comme celle de a=1,b=3)")

```

```python
visualiser_croissance(3,0.5)
print("la fonction est croissante,mais la croissance vers plus l'infinie est plus lente que dans le cas précédent")
```

```python
visualiser_croissance(1,-5)
print("La fonction decroît de manière exponentielle,la courbe descent vers 0")
```

```python
visualiser_croissance(10,-0.5)
print("La fonction decroît plus lentement vers 0 même si a=10>1")
```

```python
visualiser_croissance(5,0)
print("La fonction est constante,la courbe est une ligne horizontale à y=5")
```

```python
visualiser_croissance(-1,0.5)
print("La fonction décroît vers moins l'infinie ")

```

```python
visualiser_croissance(-1,2)
print("La fonction décroît plus rapidement vers moins l'infinie que la fonction précédente ") 
```

```python
visualiser_croissance(-2,-4)
print("La fonction croît de maniére exponentielle vers 0")
```

```python
visualiser_croissance(-5,-0.2)
print("La fonction croît en 0 plus lentement que la fonction precedente ")                    
```

# Conclusion



## L'influence combinée de a et b 

### b>0:

-Si a>0:la fonction croît vers plus l'infinie 

-Si a<0:la fonction décrôit vers moins l'infinie

### b<0:

-Si a>0:la fonction decroît vers 0

-Si a<0:la fonction croît vers 0

## Le rôle de "b":(taux de croissance/décroissance)

-Si b different de zéro ,plus la valeur absolue b est grand ,plus la croissance ou la décroissance est rapide

-si b=0,La fonction est constante

## Le rôle de "a":(amplitude et direction)

-Si a>0:

Si b > 0,la fonction crôit vers plus l'infinie

Si b < 0,la fonction décroit vers 0

-Si a<0:

Si b>0,la fonction décroît vers moins l'infinie

Si b<0,la fonction croît vers 0

-Si a=0:fonction nulle
